```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
#loc=nilai mean
```

Simple Violin Plot

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Pengaturan pada Violin Plot

```python
plt.violinplot(data,
              showextrema=True,# menunjukkan garis tengah dan nilai max dan min
              showmeans=False,
              showmedians=False,
              quantiles=[0.25,0.5,0.75])

#karena distribusi normal mean dan median ada di posisi yang sama
plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Horizontal Violin Plot

```python
plt.violinplot(data, vert=False)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Multiple Violin Plot

```python
np.random.seed(10)
data1 = np.random.normal(100,10,200)
data2 = np.random.normal(80,30,200)
data3 = np.random.normal(90,20,200)
data4 = np.random.normal(70,25,200)

data = [data1, data2, data3, data4]

plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
