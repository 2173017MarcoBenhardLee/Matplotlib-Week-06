```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Sample dataset

```python
np.random.seed(2)
# np.random.seed() = Generates pseudo-random number based on seed. Seed can be any number.
data = np.random.normal(loc=100,scale=10,size=200)
data
# loc = mendefinisikan titik tengah atau nilai mean
# scale = persebaran data / standard deviation
# size = jumlah data point yang ingin dihasilkan
```

Simple Box Plot

Garis pada box = median
Atas kotak = Q3
Bawah kotak = Q1

Garis diatas kotak = nilai tertinggi
Garis dibawah kotak = nilai terendah

lingkaran = outlier pada data

```python
plt.boxplot(data, labels=['Data'])

# plt.boxplot(data yang akan di floating, parameter labels untuk mengatur sticker atau penanda sumbu X dengan data berupa list)

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()

plt.show()
```

Notched Box Plot

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Pengaturan marker untuk outlier

```python
data = np.random.normal(loc=100,scale=10,size=200)
plt.boxplot(data,
           labels=['Data'],
           showfliers=True, #untuk menunjukkan / menghilangkan outlier
           flierprops={'markerfacecolor':'r','marker':'d'})
            # untuk mengedit outlier menggunakan dictionary 'd' = bentuk diamond, 's' = square
plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Horizontal Box Plot

```python
plt.boxplot(data, labels=['Data'], vert=False)

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

Multiple Box Plot

```python
np.random.seed(3)
data1 = np.random.normal(100,10,200)
data2 = np.random.normal(80,30,200)
data3 = np.random.normal(90,20,200)
data4 = np.random.normal(70,25,200)

data = [data1, data2, data3, data4]
# Datanya di list
labels = ['Data 1','Data 2', 'Data 3', 'Data 4']
# Labelnya di simpan dilist
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
