```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

Kasus twin axes 1

```python
x= np.linspace(1,10,25)#(nilai terkecil, terbesar, jumlah data yang akan dihasilkan)
x
```

```python
fig, ax1 = plt.subplots()
#.subplots() mengembalikan 2 object dalam bentuk variable

ax1.plot(x, np.exp(x),'bs-', label='exp') #(sumbu x, sumbu y,'bs-'=blue square line/garis,label legend)
# np. exp() calculates e^x for each value of x in your input array
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx() #.twinx akan menghasilkan axes baru yang akand ditempatkan diatas axes 1
ax2.plot(x, np.log(x),'ro-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left') #lower left, lower right, upper right

plt.show()
```

Kasus twin axes 2

```python
x = np.random.randint(16, 40, size=30)
x
```

```python
def C2F(celcius=0):
    return (celcius * 1.8) + 32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim() #ylim() = limit y
    # y1 = menampung nilai terkecil
    # y2 = menampun nilai terbesar
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()
    
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.callbacks.connect('ylim_changed', konversi_sumbu)# dengan ini terjadi sinkronisasi sumbu Y yg dimiliki X1 dengan sumbu Y yang dimiliki ax2
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celcius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperatur Udara')
plt.show()
```
